```python
from langgraph.graph import StateGraph, START,END
from langchain_google_genai import ChatGoogleGenerativeAI
from typing import TypedDict
from dotenv import load_dotenv
load_dotenv()
```

```python
model = ChatGoogleGenerativeAI(model='gemini-2.0-flash')
model.invoke("What is mL").content
```

```python
class llmState(TypedDict):
    question:str
    answer:str
```

```python
def ask_question(state: llmState)->llmState:
    question=state['question']
    prompt=f"Answer the following question in brief:\n{question}"
    state['answer']=model.invoke(prompt).content
    return state


def get_answer(state: llmState)->llmState:
    return state
```

```python
graph=StateGraph(llmState)

graph.add_node("ask_question",ask_question)
graph.add_node("get_answer",get_answer)

graph.add_edge(START,"ask_question")
graph.add_edge("ask_question","get_answer")
graph.add_edge("get_answer",END)

workflow=graph.compile()
```

```python
final_state=workflow.invoke({'question':'What is machine learning?'})
print(final_state)
```
